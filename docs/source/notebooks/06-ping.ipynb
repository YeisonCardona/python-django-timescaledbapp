```python
def generate_bytes(channels=16, rows=1):
    data = [
        # 0xa0,       # header
        # id_ % 256,  # ID 0-255
        *eeg_(channels),
        # *aux_(1),
        # 0xc0,       # footer
        ] * rows
    
    return bytes(data)
```
